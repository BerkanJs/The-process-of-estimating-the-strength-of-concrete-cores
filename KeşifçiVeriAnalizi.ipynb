#### Cement (çimento) :

##### özet olarak çimento değişkeninin standart sapması çok yüksek ortalama(281.167864) etrafında +- 104.506364 kg dağılıyor. 


```python
cv_cement=(df["cement"].std()/df["cement"].mean())*100
print(f"Coefficient of Variation : %{cv_cement} bu değer %25in üzerinde yayılmanın fazla olduğunu gösteriyor bize") 
```

#### Age (yaş):

##### Yorum: Beton yaşı büyük bir varyasyona sahip. Çoğu beton 1 hafta ile 1 ay arasında yaşlandırılmış.

#### Flyash (kül)

##### Yorum: Kül miktarı genelde sıfır veya düşük seviyelerde çoğu numunelerde kül kullanılmamış.

```python

def calculate_cv(column):
    return (df[column].std() / df[column].mean()) * 100


cv_values = {col: calculate_cv(col) for col in df.columns}


for col, cv in cv_values.items():
    print(f"Coefficient of Variation for {col}: %{cv:.2f}")
    if cv > 25:
        print(f"  -> {col} verisi geniş bir dağılıma sahip, yayılma fazla!")
    else:
        print(f"  -> {col} verisi dar bir dağılıma sahip.")
```

- Sonuç: Beton karışımındaki bileşenler, önemli bir varyasyona sahip. Çimento, cüruf ve kül miktarlarında geniş bir dağılım bulunuyor. Su, kaba ve ince agrega miktarları ise daha dar bir dağılıma sahip. Beton yaşı ve mukavemeti de geniş bir varyasyon gösteriyor. Bu durum, karışımların farklılık gösterdiğini ve beton mukavemetinin birçok faktöre bağlı olarak değiştiğini gösteriyor.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


fig, axes = plt.subplots(nrows=len(df.columns), ncols=2, figsize=(14, 4 * len(df.columns)))


for i, column in enumerate(df.columns):
   
    sns.histplot(df[column], kde=True, bins=20, ax=axes[i, 0])
    axes[i, 0].set_title(f'{column} Histogram')
    

    sns.boxplot(x=df[column], ax=axes[i, 1])
    axes[i, 1].set_title(f'{column} Boxplot')


plt.tight_layout(pad=5.0)  

plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


fig, axes = plt.subplots(nrows=len(df.columns) - 1, ncols=1, figsize=(8, 5 * (len(df.columns) - 1)))


for i, column in enumerate(df.columns):
    if column != 'csMPa':  # 
        sns.scatterplot(data=df, x=column, y='csMPa', ax=axes[i])
        axes[i].set_title(f'{column} vs csMPa')
        axes[i].set_xlabel(column)
        axes[i].set_ylabel('csMPa')


plt.tight_layout(pad=5.0)

plt.show()
```

#### Değişkenler çoğunlukla doğrusal dağılmadığı gibi çaprazlama grafiklere baktığımızda hedef bağımlı değişken ile lineer ilişki içinde de değiller.
- scatter plot'ta dikey bir çizgi görünmesi, doğrusal bir ilişki olmadığını ve korelasyonun zayıf veya yok olduğunu gösterir.

```python
import matplotlib.pyplot as plt
import seaborn as sns


min_strength = df['csMPa'].min()
max_strength = df['csMPa'].max()


min_strength_row = df[df['csMPa'] == min_strength]
max_strength_row = df[df['csMPa'] == max_strength]


print("En Düşük Beton Dayanımı (csMPa):", min_strength)
print(min_strength_row)
print("\nEn Yüksek Beton Dayanımı (csMPa):", max_strength)
print(max_strength_row)


plt.figure(figsize=(10,6))


sns.histplot(df['csMPa'], kde=True, color='blue', label='csMPa')
plt.axvline(min_strength, color='red', linestyle='--', label='En Düşük Dayanım')
plt.axvline(max_strength, color='green', linestyle='--', label='En Yüksek Dayanım')

plt.title('Beton Dayanımının (csMPa) Dağılımı\nEn Düşük ve En Yüksek Dayanım Karşılaştırması')
plt.xlabel('Beton Dayanımı (csMPa)')
plt.ylabel('Frekans')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


comparison_dict = {'5 lowest strength': df.groupby(by='csMPa').mean().iloc[0:5].mean(),
                  '5 greatest strength': df.groupby(by='csMPa').mean().iloc[-5:].mean()}


comparison_df = pd.DataFrame(comparison_dict)


comparison_df = comparison_df.T
comparison_df.columns.name = 'Features'
comparison_df.index.name = 'Concrete Strength Comparison'


styled_comparison_df = comparison_df.style.background_gradient(cmap='coolwarm', axis=None).format("{:.2f}")


styled_comparison_df

```

- Her iki grupta da kül miktarı sıfır.Bu istatistik,bu veri seti için, külün beton dayanımını etkileyen bir faktör olmadığını gösteriyor.

- Su miktarının, daha düşük beton dayanım gruplarında daha yüksek olduğunu görebiliriz. Bu, su ve çimento oranının betonun dayanımını etkileyen bir faktör olabileceğini düşündürür. Genelde daha az su, daha sağlam bir beton sağlar.

- Çimento ve su miktarı daha yüksek beton dayanımlarında artarken, süperplastikleştirici daha düşük dayanımlar için yok denecek kadar az kullanılmış. Bu, bazı malzemelerin beton dayanımı üzerinde daha önemli bir etkisi olduğunu gösteriyor.

- Yaşın beton dayanımını önemli ölçüde artırdığı açıkça görülüyor; beton ne kadar eskiyse, dayanım o kadar yüksek olabiliyor.
