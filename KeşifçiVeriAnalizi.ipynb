#### Age (yaÅŸ):

##### Yorum: Beton yaÅŸÄ± bÃ¼yÃ¼k bir varyasyona sahip. Ã‡oÄŸu beton 1 hafta ile 1 ay arasÄ±nda yaÅŸlandÄ±rÄ±lmÄ±ÅŸ.

#### Flyash (kÃ¼l)

##### Yorum: KÃ¼l miktarÄ± genelde sÄ±fÄ±r veya dÃ¼ÅŸÃ¼k seviyelerde Ã§oÄŸu numunelerde kÃ¼l kullanÄ±lmamÄ±ÅŸ.

```python

def calculate_cv(column):
    return (df[column].std() / df[column].mean()) * 100


cv_values = {col: calculate_cv(col) for col in df.columns}


for col, cv in cv_values.items():
    print(f"Coefficient of Variation for {col}: %{cv:.2f}")
    if cv > 25:
        print(f"  -> {col} verisi geniÅŸ bir daÄŸÄ±lÄ±ma sahip, yayÄ±lma fazla!")
    else:
        print(f"  -> {col} verisi dar bir daÄŸÄ±lÄ±ma sahip.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


fig, axes = plt.subplots(nrows=len(df.columns), ncols=2, figsize=(14, 4 * len(df.columns)))


for i, column in enumerate(df.columns):
   
    sns.histplot(df[column], kde=True, bins=20, ax=axes[i, 0])
    axes[i, 0].set_title(f'{column} Histogram')
    

    sns.boxplot(x=df[column], ax=axes[i, 1])
    axes[i, 1].set_title(f'{column} Boxplot')


plt.tight_layout(pad=5.0)  

plt.show()


```

```python
import matplotlib.pyplot as plt
import seaborn as sns


min_strength = df['csMPa'].min()
max_strength = df['csMPa'].max()


min_strength_row = df[df['csMPa'] == min_strength]
max_strength_row = df[df['csMPa'] == max_strength]


print("En DÃ¼ÅŸÃ¼k Beton DayanÄ±mÄ± (csMPa):", min_strength)
print(min_strength_row)
print("\nEn YÃ¼ksek Beton DayanÄ±mÄ± (csMPa):", max_strength)
print(max_strength_row)


plt.figure(figsize=(10,6))


sns.histplot(df['csMPa'], kde=True, color='blue', label='csMPa')
plt.axvline(min_strength, color='red', linestyle='--', label='En DÃ¼ÅŸÃ¼k DayanÄ±m')
plt.axvline(max_strength, color='green', linestyle='--', label='En YÃ¼ksek DayanÄ±m')

plt.title('Beton DayanÄ±mÄ±nÄ±n (csMPa) DaÄŸÄ±lÄ±mÄ±\nEn DÃ¼ÅŸÃ¼k ve En YÃ¼ksek DayanÄ±m KarÅŸÄ±laÅŸtÄ±rmasÄ±')
plt.xlabel('Beton DayanÄ±mÄ± (csMPa)')
plt.ylabel('Frekans')
plt.legend()
plt.show()
```
